# Comprehension 내포

- Comprehension은 한 줄짜리 반복문으로 새로운 컬렉션을 만드는 문법이다.
- 대표적으로 list comprehension과 dict comprehension이 있다.
- 반복문으로 값을 하나씩 추가하는 패턴을 더 짧고 읽기 좋게 표현할 수 있다.

- list comprehension
```
# list 생성
[표현식 for 변수 in 반복가능한객체]

# 조건 포함
[표현식 for 변수 in 반복가능한객체 if 조건식]
```

- dict comprehension
```
# dict 생성
{key표현식: value표현식 for 변수 in 반복가능한객체}

# 조건 포함
{key표현식: value표현식 for 변수 in 반복가능한객체 if 조건식}
```



## 1. for 반복문과 list comprehension 비교

먼저 일반 `for` 반복문으로 리스트를 만든 뒤, 같은 결과를 list comprehension으로 바꿔 본다.

````python
def make_numbers_with_for() -> list[int]:
    numbers: list[int] = []

    for n in range(1, 6):
        numbers.append(n)
    return numbers

print(make_numbers_with_for())

# list comprehension

def make_numbers_with_comprehension() -> list[int]:
    return [n*2 for n in range(1,6)]


print(make_numbers_with_comprehension())
````

## 2. 조건식이 있는 list comprehension

`if` 조건을 붙이면 원하는 요소만 필터링해서 리스트를 만들 수 있다.

````python
# 1 ~ 100 사이 정수 중 3의 배수만 골라서 list 요소로 추가
def make_numbers3() -> list [int]:
    return [n for n in range(1, 101) if n % 3== 0]

print("3의 배수 리스트:  ", make_numbers3())
````

````python
# 전달된 list에서 정수(int) 타입만 필터링해서 반환
lst = [1, '123', '안녕', 22, {}, [],True, 333 ]

#object 타입: 아무 타입이나 가능 (==Any)
def filter_int_values(values: list[object]) -> list [int]:
    # isinstance(value, type) -> bool
    # - value의 자료형이 type과 일치하면 True, 아니면 False
    return [v for v in values if isinstance(v,int)]

print(filter_int_values(lst)) # True == 1
````

## 3. 중첩 반복문 list comprehension

반복문이 중첩된 구조도 comprehension으로 표현할 수 있다. 다만 너무 복잡해지면 일반 반복문이 더 읽기 쉽다.

````python
# 중첩 반복문
def make_positions() -> list[tuple[int,int]]:
    positions: list[tuple[int,int]] = []

    for row in range(3): #0, 1, 2
        for col in range(5): # 0, 1, 2, 3, 4
            positions.append((row, col))

    return positions

print(make_positions())
````

````python
# 중첩 comprehension
def make_positions_with_comprehensions() -> list[tuple[int, int]]:
    return [(row,col)for row in range(3) for col in range(5)]

print(make_positions_with_comprehensions())
````

## 4. dict comprehension

`dict`도 comprehension으로 만들 수 있다.

```python
{key식: value식 for 요소 in iterable}
```

````python
def make_square_map() -> dict[int, int]:
    return {k: k **2 for k in range(1,6)}

print (make_square_map())
````

````python
# 문자열 리스트를 전달하여 {"문자열" : 문자열 길이} 형태의 list[dict] 반환
def make_word_length_map(words: list[str]) -> dict[str, int]:

    # 일반적인 방법
    # result = {}
    # for word in words:
    #     result[word] = len(word)
    # return result

    #comprehension
    return  {word: len(word)  for word in words if len(word) > 3}

words = ['hello', '배고파', '반복문은 왜 이렇게 어려운가']
print (make_word_length_map(words))
````

## 5. list와 dict 변환

`zip()`과 `dict()`를 함께 사용하면 두 리스트를 하나의 딕셔너리로 묶을 수 있다. 반대로 `dict.items()`와 unpacking을 사용하면 딕셔너리에서 key와 value 묶음을 분리할 수 있다.

````python
def make_age_map() -> dict[str, int]:
    names: list[str] = ['홍길동', '신사임당', '이순신']
    ages: list[int] = [20, 40, 30]

    # zip (names, ages == ['홍길동', 20, '신사임당',  40, '이순신', 30]
    return dict(zip(names, ages))

print("list -> dict", make_age_map())
````

````python
def split_age_map() ->tuple[tuple[str, ...], tuple[int, ...]]:
    info: dict[str, int] = {'홍길동': 20, '신사임당': 40, '이순신': 30}
    print("info.items(): ", info.items()) #list[tuple]
    print(*info.items()) # * 기호 == 풀기

    names, ages = zip(*info.items())
    return names, ages

print("dict-> tuple묶음:",split_age_map())
````
