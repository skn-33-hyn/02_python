# 추상클래스 Abstract Class
- 자식클래스에서 구현해야 할 추상메소드를 가진 부모클래스 (인터페이스)
   - 추상 메서드 == 미완성 메서드 == 메서드 정의 부분이 비어있음

- 추상메서드는 자식클래스에서 반드시 구현(오버라이딩)해야할 의무를 가진다.

- 추상클래스는 직접 객체화 할수 없다. (추상클래스로 인스턴스 생성 불가)

- abc.ABC클래스를 상속하면 추상클래스를 선언할 수 있다.
    - ABC == Abstract Base Class
- @abstractmethod 데코레이터로 추상메소드를 선언할 수 있다.

```python
from abc import ABC,  abstractmethod

#추상 클래스 정의
# - ABC 상속
# == 해당 클래스가 너무 추상적 또는 해당 클래스로 인스턴스 생성 X

# 추상 클래스 목적
# - 상속받은 자식 클래스들에게 공통된 규약(규칙)을 제공하는 용도
class Product(ABC):

    def __init__(self, code: str, name: str, price: int):
        self.__code = code
        self.__name = name
        self.__price = price

    # 추상 메서드 == 미완성 메서드 == 정의 부분이 없음
    @abstractmethod # 추상 메서드임을 표시
    def get_tax(self):
        pass # 정의 부분을 상속 받은 자식에게 넘김

    def get_price(self):
        return self.__price

    def __str__(self):
        return f'code={self.__code}, name={self.__name}, price={self.__price}'
```

```python
# 추상 클래스로 인스턴스 생성하기 -> 오류
# prod = Product("111","노트북", "200000")
# TypeError: Can't instantiate abstract class
```

```python
# 추상 클래스를 상속 받은 자식 클래스
class TV(Product):

    def __init__(self, code: str, name: str, price: int,size : int):
        super().__init__(code, name, price)
        self.__size = size

    #상속 받은 추상 메서드를 재정의 == 오버라이딩
    def get_tax(self): #부과세를 10%로 책정해서 반환
        return self.get_price() * 0.1

    # __str__()오버라이딩
    def __str__(self):
        return f'{super().__str__()} size={self.__size}'
```

```python
tv1 =TV("samsung-tv-1234","SAMSUNG OLED TV",5_000_000, 85)
print(tv1)
print(tv1.get_tax())

tv2 =TV("LG-tv-0001","LG Flexible TV",10_000_000, 100)
print(tv2)
print(tv2.get_tax())
```
